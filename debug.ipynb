```python
import glob
import json
from multiprocessing import Pool
from tqdm import tqdm

def findher(filepath: str):
    with open(filepath, "r") as f:
        for line in f:
            if "274623" in line:
                paper = json.loads(line)
                if paper['corpusid'] == 274623:
                    return (filepath, paper)
    return None


filepaths = list(glob.glob("data/raw/s2orc/*"))
candidates = []
with Pool(10) as p:
    for result in tqdm(p.imap(findher, filepaths)):
        if result is not None:
            break
```
